# 準備工作

```python
# 準備工作

from bs4 import BeautifulSoup
import time
import datetime
import re
import sqlite3
import os
import json
import csv
import pprint

data_source = "../data/台北市停車場.json"
output_file_path = "../data/台北市停車場_filtered.csv"
```

```python
# 讀取全臺縣市區域街道查詢表

addr_source = "../data/全台鄉鎮市區街道列表.csv"

with open(addr_source) as addr_file:    
    addr_ref = csv.reader(addr_file)
    addr_ref = [tuple(row) for row in addr_ref]
```

# 定義方法

```python
def get_each_result_list(each_dict):

#     pprint.pprint(each_dict)

    each_coord = each_dict['Entrancecoord']['EntrancecoordInfo'][0]
#         print(each_coord)

    name = each_dict['name']

    # Entrancecoord

    try:
        address = each_dict['address']
    except:
        address = each_dict['addresss']
    
    lat = float(each_coord['Xcod'])
    lng = float(each_coord['Ycod'])

    # results

    each_result_list = [name,address,lat,lng]        
    return each_result_list
```

# 執行程式

```python
# 讀取已存好的資料

with open(data_source,encoding='big5') as data_file:    
    main_data = json.load(data_file)
```

```python
pprint.pprint(main_data)
```

```python
# Main loop

main_data_park = main_data['data']['park']
filtered_table = []

for each_dict in main_data_park:
    try:
        filtered_table.append(get_each_result_list(each_dict))
    except Exception as e:
        print("Exception: " + str(e))
```

```python
pprint.pprint(filtered_table)
```

```python
with open(output_file_path,'wt') as dest_file:    
    output = csv.writer(dest_file, delimiter=',')
    output.writerows(filtered_table)

print("CSV file saved.")
```
