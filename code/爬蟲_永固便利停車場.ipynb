```python
# 準備工作

from bs4 import BeautifulSoup
import requests
import bs4
import time
import re
import sqlite3
import pprint
```

```python
# 定義 cleanse 方法
# 把爬下來的字裡面不必要的字元清除掉

def cleanse(string):
    
    string = string.replace("\u3000","")
    string = string.replace("\r","")
    string = string.replace("\n","")
    string = string.replace(" ","")
    string = string.replace("、",",")
    string = string.replace("，",",")
    
    return string
```

```python
def crawl(number, dictionary):
    
    temp = {}
    
    ID = format(number, '05')    # 補齊ID前面的0
    href = "http://www.24tps.com.tw/ParkMap/Map.aspx?ID=AA" + str(ID)
    res = requests.get(href)
    soup = BeautifulSoup(res.text,'html.parser')

    td = soup.select('td[style="vertical-align: top; width: 375px; height: 325px; text-align: left; background-color: #ecfff1; color: chocolate;"]')
    
    # 站名
    name = td[0].select('td')[2].text

    # 地址
    address = td[0].select('td')[4].text

    # 電話
    phone_number = td[0].select('td')[6].text

    # 營業時間
    business_hours = td[0].select('td')[8].text

    # 車高限制
    height_limit = td[0].select('td')[10].text

    # 停車費率
    fee = td[0].select('td')[12].text

    # 營業性質
    services = td[0].select('td')[14].text

    # 請洽
    contacts = td[0].select('td')[16].text

    # 附註
    notes = td[0].select('td')[17].text
    
    temp = {\
             '站名':name,\
             '地址':address,\
             '電話':phone_number,\
             '營業時間':business_hours,\
             '車高限制':height_limit,\
             '停車費率':fee,\
             '營業性質':services,\
             '請洽':contacts,\
             '附註':notes\
           }
           
    for key in temp:
        value = cleanse(temp[key])
        temp.update({key:value})
    
    dictionary.update({ID:temp})
    
```

```python
parking_lot_dict = {}

for number in range(1,200):
    try:        
        crawl(number, parking_lot_dict) 
        print(parking_lot_dict[format(number, '05')])
    except:        
        print("發生例外,可能是網頁不存在") 
    finally:
        print(number)
```

```python
# 建立sqlite資料庫儲存

# ***請在當下資料夾沒有24tps_parking_data的時候執行

db_name = "../data/24tps_parking_data.sqlite"

with sqlite3.connect(db_name) as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE lots(
    ID TEXT NOT NULL UNIQUE,
    站名 TEXT NOT NULL,
    地址 TEXT,
    電話 TEXT,
    營業時間 TEXT,
    車高限制 TEXT,
    停車費率 TEXT,
    營業性質 TEXT,
    請洽 TEXT,
    附註 TEXT)""")
            
conn.close()
```

```python
# 儲存檔案為sqlite檔

db_name = "../data/24tps_parking_data.sqlite"

columns = ["站名", "地址", "電話", "營業時間", "車高限制", "停車費率", "營業性質", "請洽", "附註"]
columns_string = "ID, " + ",".join(columns)

with sqlite3.connect(db_name) as conn2:
    
    c2 = conn2.cursor()

    for key in parking_lot_dict:

        values = str(key)

        for each in columns:
            values += ",'" + parking_lot_dict[key][each] + "'"            
        
        insertString = "INSERT INTO lots (" + columns_string + ") VALUES(" + values + ");"
        c2.execute(insertString)

conn2.close()
```

```python
# 檢查結果

with sqlite3.connect(db_name) as conn3:
    
    c3 = conn3.cursor()
    qryString = "SELECT * from lots;"
    c3.execute(qryString)
    results = c3.fetchall()

print(results)
```
